````python
from ats_keywords import ATSKeywords
import logging

# Configure logging
logging.basicConfig(
    filename='prompt_testing.log',  # File to append the logs
    level=logging.DEBUG,  # Log level set to DEBUG
    format='%(asctime)s - %(levelname)s - %(message)s',  # Log format
)

def test_prompt(_extract_keywords_prompt, _test_postings, model = "gpt-4"):
    for _posting in _test_postings:
        text = _posting.lstrip().replace("\n\n\n\n", "\n\n")
    
        # Split the text by lines
        lines = text.split("\n")
        
        # Remove leading whitespaces from each line
        trimmed_lines = [line.lstrip() for line in lines]
        
        # Join the lines back into a single string
        trimmed_text = "\n".join(trimmed_lines)

        kw_extractor = ATSKeywords()
        test_output = kw_extractor.extract_keywords_from_posting(_extract_keywords_prompt, trimmed_text, model=model)

        logging.debug(f"Prompt: {_extract_keywords_prompt.format(job_posting_content=_posting)}")
        logging.debug(f"Output: {test_output}")
        logging.debug("=" * 50)

        print(trimmed_text)
        print(test_output)
        print("=" * 50)
````
